```python
import cv2
import matplotlib.pyplot as plt
import os

# 1. Path-г тусдаа хувьсагчид хадгалъя
path = '..\data\external\sample.jpg'

# 2. Файл байгааг шалга
print("File exists:", os.path.exists(path))

# 3. Зураг унших
img = cv2.imread(path)

# 4. Хэрвээ зураг олдоогүй бол алдаа гаргах
if img is None:
    print("❌ Зураг олдсонгүй. Path шалгана уу:", path)
else:
    # 5. BGR → RGB хөрвүүлэлт
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 6. Зураг харуулах
    plt.imshow(img_rgb)
    plt.title("Sample Image")
    plt.axis('off')
    plt.show()
```
