```python
import cv2
import matplotlib.pyplot as plt
import os

# 1. Path-   
path = '..\data\external\sample.jpg'

# 2.   
print("File exists:", os.path.exists(path))

# 3.  
img = cv2.imread(path)

# 4.      
if img is None:
    print("  . Path  :", path)
else:
    # 5. BGR  RGB 
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 6.  
    plt.imshow(img_rgb)
    plt.title("Sample Image")
    plt.axis('off')
    plt.show()
```
